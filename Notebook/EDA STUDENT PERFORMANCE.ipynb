```python
df['total score']= df['math_score'] + df['reading_score'] + df['writing_score']
df['average']= df['total score']/3
df.head()
```

```python
numerical_feature=[feature for feature in df.columns if df[feature].dtype != 'O']
catergorical_feature= [feature for feature in df.columns if df[feature].dtype =='O']

print('we have toral {} number of numerical features :{}'.format(len(numerical_feature),numerical_feature))
print('\nwe have {} number of categorical features :{}'.format(len(catergorical_feature), catergorical_feature))

```

```python
reading_full= df[df['reading_score']==100]['average'].count()
writing_full= df[df['writing_score']==100]['average'].count()
math_full= df[df['math_score']==100]['average'].count()



print('Total number od student with full marks in reading :',math_full)
print('Total number of student with full marks in writing:',writing_full)
print('Total number of student with full maeks in reading:', writing_full)


```

```python
low_reading= df[df['reading_score']<=20]['average'].count()
low_writing= df[df['writing_score']<=20]['average'].count()
low_math= df[df['math_score']<=20]['average'].count()

print('Total number of students with lower reading_scores:', low_reading )
print('Total number of students with lower writing_scores: ',low_writing)
print('Total number of stduents with lower marks in math: ',low_math)
```

Multivariate Analysis using pieplot


```python
plt.figure(figsize=(25,6))

plt.subplot(1,5,1)
size=df['gender'].value_counts()
label='male','female'
color= ['red','green']

plt.pie(size,colors=color,labels=label,autopct='.%2f%%')
plt.title('Gender',fontsize=20)
plt.axis('off')


plt.subplot(1,5,2)
size=df['race_ethnicity'].value_counts()
label='group A','group B','group C','group D','group E'
color=['RED','Blue','Green','Yellow','Brown']

plt.pie(size,colors=color,labels=label, autopct='.%2f%%')
plt.title('race_ethnicity')
plt.axis('off')



plt.subplot(1,5,3)
size=df['parental_level_of_education'].value_counts()
label='some college',"associate's degree",'high school','some high school',"bachelor's degree","master's degree"
color=['Red','Blue','Green','Yellow','Brown','Grey']

plt.pie(size,colors=color,labels=label,autopct='.%2f%%')
plt.title('parental_level_of_education')
plt.axis('off')



plt.subplot(1,5,4)
size=df['lunch'].value_counts()
label='standard','free/reduced'
color=['Red','yellow']

plt.pie(size, colors= color, labels= label, autopct='.%2f%%')
plt.title('Lunch')
plt.axis('off')

plt.subplot(1,5,5)
size=df['test_preparation_course'].value_counts()
label='None','Completed'
color=['Red','Yellow']

plt.pie(size,labels=label,colors=color,autopct=".%2f%%")
plt.title('test_preparation_course')
plt.axis('off')


plt.tight_layout()
plt.grid()

plt.show()

```

```python
f,ax=plt.subplots(1,2,figsize=(20,10))
sns.countplot(x=df['gender'],data=df,palette ='bright',ax=ax[0],saturation=0.95)
for container in ax[0].containers:
    ax[0].bar_label(container,color='black',size=20)
    
plt.pie(x=df['gender'].value_counts(),labels=['Male','Female'],explode=[0,0.1],autopct='%1.1f%%',shadow=True,colors=['#ff4d4d','#ff8000'])
plt.show()

```

```python
gender_group = df.groupby('gender').mean()
gender_group
```

```python
plt.figure(figsize=(10, 8))

X = ['Total Average','Math Average']


female_scores = [gender_group['average'][0], gender_group['math_score'][0]]
male_scores = [gender_group['average'][1], gender_group['math_score'][1]]

X_axis = np.arange(len(X))
  
plt.bar(X_axis - 0.2, male_scores, 0.4, label = 'Male')
plt.bar(X_axis + 0.2, female_scores, 0.4, label = 'Female')
  
plt.xticks(X_axis, X)
plt.ylabel("Marks")
plt.title("Total average v/s Math average marks of both the genders", fontweight='bold')
plt.legend()
plt.show()
```
